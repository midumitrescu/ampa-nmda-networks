We use the states that Brunnel defined:
1. SR - neurons are almost fully synchronized in very few clusters. Behave as oscillators. Excitation dominates inhibition. Mean driven regime => mean is above 0, which is used as threshold
2. SI fast - oscillatory global activity but individual neurons are firing irregularly. Stationary global activity (?) but strongly irregular individual firing (oscillations, but CV is 1 for individual neurons)
3. AI - stationary global activity but strongly irregular (population firing rate constant, CV = 1). Firing at low rates. Inhibition dominates
4. SI slow
5. Q  - Quiescent state. Very low activity
