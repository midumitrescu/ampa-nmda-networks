There are two variables and 4 parameters controlling the evolution of NMDA channels:

1. $\frac{dx}{dt} = \underbrace{\phi}_{\text{scalar}} \left( \alpha_x \cdot \sum_j \delta(t-t_j) - \frac{x}{\tau_x}\right)$
2. $\frac{ds}{dt} = \phi \left( \alpha_s \cdot x \cdot \underbrace{(1-s)}_{\text{closed/unsaturated/available NMNDA channels}}  - \frac{s}{\tau_s}\right)$
3. $I_\text{NMDA} = g_\text{NMDA} \cdot \underbrace{s}_{\text{how much NMDA is open}} \cdot (V - E_E) \cdot \frac{1}{1 + \frac{\left[ \text{Mg}^{2+} \right]}{3.57} \exp{(-0.062 V)}}$


The $0.062 V$ suggests the correct magnitude of the membrane voltage is millivolts. The factor $\frac{\left[ \text{Mg}^{2+} \right]}{3.57}$ scales the sigmoid, with the effect of moving the typical $1/2$ point. In our case, it moves it "up". I.e.

## When is most of the magnesium unblocked?

We start with the voltage-dependent magnesium block equation for NMDA receptors:

$\sigma(V) = \frac{1}{1 + [\text{Mg}^{2+}] \cdot \exp(-0.062 V)/3.57}$

Assuming normal extracellular magnesium concentration:
$[\text{Mg}^{2+}] = 1 \, \text{mM}$

This simplifies the equation to:
$\sigma(V) = \frac{1}{1 + \exp(-0.062 V)/3.57}$

We want to find the membrane potential \( V \) at which **at least 75%** of the NMDA current is unblocked:
$\sigma(V) \geq 0.75$

$\frac{1}{1 + \exp(-0.062 V)/3.57} \geq 0.75$

$1 + \frac{\exp(-0.062 V)}{3.57} \leq \frac{1}{0.75} = \frac{4}{3}$

$\frac{\exp(-0.062 V)}{3.57} \leq \frac{1}{3}$

$\exp(-0.062 V) \leq 1.19$

$-0.062 V \leq \ln(1.19) \approx 0.174$

$V \geq \frac{-0.174}{-0.062} \approx 2.8 \, \text{mV}$


### Conclusion:

More than **75% of the NMDA current is unblocked** when the membrane voltage is above:

$\boxed{V \geq -3 \, \text{mV}}$

This reflects the fact that NMDA receptors require **postsynaptic depolarization** to relieve the magnesium block and allow current to flow.
